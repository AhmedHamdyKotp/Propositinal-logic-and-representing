```python
all_posibles = []
rep = [(M,S,H,W,K1,K2) for M in [True , False]
       for S in [True , False] for  H in [True , False]
      for  W in [True , False] for K1 in [True , False]
      for  K2 in [True , False]]
for(M,S,H,W,K1,K2) in rep :
  all_posibles.append((M,S,H,W,K1,K2))
print (all_posibles)

```

```python
import pandas as pd
cols = ['M','S','H','W','K1','K2']
dp = pd.DataFrame(all_posibles,columns = cols)
```

```python
dp
```

```python
Trues = []

for (m, s, h, w, k1, k2) in rep:
    A = m and s
    B = h or w
    C = k1 ^ k2
    all = A and B and C
    if  all :
      Trues.append((m, s, h, w, k1, k2, A, B, C,all))
```

```python
Trues
```

```python
g = pd.DataFrame(Trues,columns = cols + ['A','B','C','all'])
```
